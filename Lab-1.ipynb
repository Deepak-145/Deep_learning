Create two 2D tensors:

```python
import tensorflow as tf

a = tf.constant([
    [1,2],
    [3,4]
])

b = tf.constant([
    [5,6],
    [7,8]
])
```

```python
tf.add(a, b)
```

```python
tf.subtract(a, b)
```

```python
tf.multiply(a, b)
```

```python
tf.divide(a, b)
```

Create two 3D tensors:


```python
a = tf.constant([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])

b = tf.constant([
    [[1,1],[1,1]],
    [[2,2],[2,2]]
])
```

```python
tf.add(a, b)

```

```python
tf.subtract(a, b)
```

```python
tf.multiply(a, b)
```

```python
tf.divide(a, b)
```
